```python
# Split the data set into training and test set in the ratio of 70:30 (Training:Test)
# capture the target column ("status") into separate vectors for training set and test set

X = parkinson.drop(["name", "status"], axis=1)

y = parkinson.pop("status")
```

```python
# scale the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler((-1,1))
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# splitting data into training and test set for independent attributes
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=.30, random_state=1)
```

```python
# Create the model using “entropy” method of reducing the entropy and fit it to training data. (5 points)
from sklearn.tree import DecisionTreeClassifier                                  
dt_model = DecisionTreeClassifier(criterion = 'entropy' )
```

```python
dt_model.fit(X_train, y_train)
```

```python
dt_model.score(X_test, y_test) 
```

```python
### Test accuracy is 86.44 %
```

```python
predicted_score = dt_model.predict(X_test)
predicted_score
```

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_test,predicted_score)
conf_mat
```

```python
sns.heatmap(conf_mat, annot=True)
plt.xlabel('Actual data')
plt.ylabel('Predicted')
```

```python
# Use regularization parameters of max_depth, min_sample_leaf to recreate the model. 
# What is the impact on the model accuracy? How does regularization help
dt_model_regularization = DecisionTreeClassifier(criterion = 'entropy', class_weight={0:.5,1:.5}, max_depth = 5, min_samples_leaf=5 )
dt_model_regularization.fit(X_train, y_train)
```

```python
dt_model_regularization.score(X_test, y_test) 
```

```python
# 1) From above regularization, the score increased to 91.5 % from earlier 86.44 %
# 2) Following are the ways the regularization work:
'''
max_depth
the first hyperparameter to tune in a Decision Tree is max_depth
It indicates how deep the decision tree can be
The deeper the tree, the more splits it has and it captures more information about data
However in general a decision tree overfits for large depth values
The tree perfectly predicts all of the train data, however, it fails to generalize the findings for new data
Hence optimum value of max_depth should be used. 
Even the lesser value of max_depth could result in underfitting

min_samples_leaf
A leaf node is a node without any children(without any further splits)
min_samples_leaf is the minimum number of samples required to be at a leaf node
min_samples_leaf guarantees a minimum number of samples in a leaf

'''

```

```python
# importance of features in the tree building ( The importance of a feature is computed as the 
#(normalized) total reduction of the criterion brought by that feature. It is also known as the Gini importance )

print(pd.DataFrame(dt_model.feature_importances_, columns = ["Imp"], index = X_train.columns))
```

```python
# Next implement the decision tree using Random Forest. 
# What is the optimal number of trees that gives the best result?

from sklearn.ensemble import RandomForestClassifier
rfcl = RandomForestClassifier(n_estimators = 41)
rfcl = rfcl.fit(X_train, y_train)
```

```python
test_pred = rfcl.predict(X_test)
rfcl.score(X_test , y_test)
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test, test_pred))
```
