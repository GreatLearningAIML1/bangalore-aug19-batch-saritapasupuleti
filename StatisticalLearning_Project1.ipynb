```python
#Write a function with name â€œGoal_diff_countâ€ which should return all the teams with their Goal Differences.
#Goal_diff_count = GoalsFor - GoalsAgainst
laliga = pd.read_csv('/Users/srao/Documents/AIML/Project1/Laliga.csv')
laliga.replace('-', 0,inplace=True)

def Goal_diff_count():  
    global laliga
    goal_diff_max = 0
    goal_diff_min = 0
    for column in laliga[['Team', 'GoalsFor', 'GoalsAgainst']]:
        columnSeriesObj = laliga[column]
        if column == 'GoalsFor':
            columnSeriesGoalsFor = laliga[column]
        elif column == 'GoalsAgainst':
            columnSeriesGoalsAgainst = laliga[column]
        else:
            columnSeriesTeam = laliga[column]
    print('The List of GoalsFor : ', columnSeriesGoalsFor.values)
    print('\n')
    print('The List of GoalsAgainst : ', columnSeriesGoalsAgainst.values)
    print('\n')
    print('The List of Teams : ', columnSeriesTeam.values)
    print('\n')
    # Now Iterate over the individual GoalsFor and GoalsAgainst columns and find diff
    for (a, b, c) in zip(columnSeriesGoalsFor.values, columnSeriesGoalsAgainst.values, columnSeriesTeam.values):
        goal_diff = int(a) - int(b)
        print('The Goal Diff count for Team %s is %s' %(c, goal_diff))
        if goal_diff >= goal_diff_max:
            goal_diff_max = goal_diff
        elif goal_diff < goal_diff_min:
            goal_diff_min = goal_diff
    print('\n')        
    print('The max goal diff is %s' %goal_diff_max)
    print('The min goal diff is %s' %goal_diff_min)

Goal_diff_count()    

    
```

```python
# Create a new column with name â€œWinning Percentâ€ and append it to the data set 
# Percentage of Winning = (GamesWon / GamesPlayed)*100

laliga = pd.read_csv('/Users/srao/Documents/AIML/Project1/Laliga.csv')
laliga.replace('-', 0,inplace=True)

def winning_chance():  
    global laliga
    list_Winning_chance = []
    for column in laliga[['Team', 'GamesWon', 'GamesPlayed']]:
        columnSeriesObj = laliga[column]
        if column == 'GamesWon':
            columnSeriesGamesWon = laliga[column]
        elif column == 'GamesPlayed':
            columnSeriesGamesPlayed = laliga[column]
        else:
            columnSeriesTeam = laliga[column]
        
    print('The List of GamesWon : ', columnSeriesGamesWon.values)
    print('\n')
    print('The List of GamesPlayed : ', columnSeriesGamesPlayed.values)
    print('\n')
    print('The List of Teams : ', columnSeriesTeam.values)
    print('\n')
    
    
    # Now Iterate over the individual GoalsFor and GoalsAgainst columns and find diff
    for (a, b, c) in zip(columnSeriesGamesWon.values, columnSeriesGamesPlayed.values, columnSeriesTeam.values):
        if b == 0:
            winning_percent = 0
        else:
            winning_percent = int(a)/int(b)*100
        list_Winning_chance.append(winning_percent)
        print('The Percent of Winning of Team %s is %s' %(c, winning_percent))
        
    print(list_Winning_chance)
    laliga['WinningChance'] = list_Winning_chance
    print(laliga)
    
winning_chance()
```
